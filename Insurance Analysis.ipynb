```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
reg=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

# Age and Charges analysis with test size 20%

```python
X_train,X_test,y_train,y_test=train_test_split(age,charges,test_size=0.2,random_state=1)
```

```python
plt.scatter(age,charges)
```

```python
X_train=X_train.values.reshape(-1,1)
age_model=reg.fit(X_train,y_train)
```

```python
X_test=X_test.values.reshape(-1,1)
```

```python
age_model.score(X_test,y_test)
```

```python
age_model.intercept_
```

```python
age_model.coef_
```

```python
predictions=age_model.predict(X_train)
```

```python
plt.scatter(age,charges)
plt.plot(X_train,predictions,color='r')
```

```python
np.sqrt(mean_squared_error(y_train,predictions))
```

```python
r2_score(y_train,predictions)
```

# Multilinear analysis with replace the categorical data into numerical data

```python
data.head()
```

```python
data.sex=data.sex.replace('female',0)
data.sex=data.sex.replace('male',1)
```

```python
data.head()
```

```python
data.smoker=data.smoker.replace('yes',1)
data.smoker=data.smoker.replace('no',0)
```

```python
data.head()
```

```python
data['region'].value_counts()
```

```python
data.region=data.region.replace('southeast',1)
data.region=data.region.replace('southwest',2)
data.region=data.region.replace('northeast',3)
data.region=data.region.replace('northwest',4)
```

```python
data.head()
```

```python
x=data[['age','bmi','smoker','children']]
```

```python
y=data['charges']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model=reg.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_train_predictions=model.predict(X_train)
```

```python
np.sqrt(mean_squared_error(y_train,y_train_predictions))
```
